```python
import pandas as pd
from pycaret.classification import *
```

## Load Data

```python
dataset=pd.read_csv('Heart.csv')
```

```python
dataset.head()
#AHD =target=y
```

```python
dataset.dtypes
```

```python
remove=['Unnamed: 0','Oldpeak']
dataset1=dataset.drop(remove,axis=1)
```

```python
dataset1.head()
```

## Training Model

```python
heart_features=['Age','Sex','ChestPain','RestBP','Chol','MaxHR','Fbs']
```

```python
experiment=setup(dataset1,target='AHD',categorical_features=heart_features)
#Since the data do not have categorical columns
```

```python
#setup= This function initializes the training environment and creates the transformation pipeline
```

```python
best_model=compare_models()
#it will highlight best performing metrics
#Here it is Ridge Classifier
```

## Testing Model

```python
# make predictions
predict_model(best_model,dataset1.tail())
#also can be runned without specifying dataset  
```

```python
#AHD =actual value
#prediction_label=predicted value
```

## Save Model

```python
save_model(best_model,model_name='Ridge-Model')
```
